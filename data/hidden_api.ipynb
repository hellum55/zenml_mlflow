```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36'
}
```

```python
url = 'https://api.boliga.dk/api/v2/sold/search/results?propertyType=3&municipality=751&salesDateMin=2010&salesDateMax=today&searchTab=1&sort=date-d&saleType=1&page=1&priceMin=1750000'
```

```python
import requests
import time

responses = []
for page in range(1, 375):  # Loop through pages 1 to 200
    paginated_url = url.replace('page=1', f'page={page}')
    r = requests.get(paginated_url, headers=headers)
    if r.status_code == 429:  # Too Many Requests
        print(f"Rate limit hit on page {page}. Retrying after a delay...")
        time.sleep(10)  # Wait for 5 seconds before retrying
        r = requests.get(paginated_url, headers=headers)  # Retry the request
    if r.status_code == 200:  # Only append if the response is successful
        responses.append(r.json())
    else:
        print(f"Failed to fetch page {page}: {r.status_code}")
r.json()
```

```python
import pandas as pd

# Extract 'results' from each response and flatten into a single list
all_results = [item for response in responses for item in response['results']]

# Convert the list of results into a DataFrame
df = pd.DataFrame(all_results)

# Display the first few rows of the DataFrame
df.head()
```

```python
df = df[['estateId', 'address', 'price', 'soldDate', 'propertyType', 
         'sqmPrice', 'rooms', 'size', 'buildYear', 
         'latitude', 'longitude', 'city']]
df.head()
```

```python
df['soldDate'] = pd.to_datetime(df['soldDate'])
df['soldDate'] = df['soldDate'].dt.date
df.head()
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['sqmPrice'] = pd.to_numeric(df['sqmPrice'], errors='coerce')
df.dtypes
```

```python
df['sqmPrice'] = df['sqmPrice'].astype(int)
df.head()
```

```python
df.describe()
```

```python
df = df[df['sqmPrice'] >= 10000]
df = df[df['sqmPrice'] <= 80000]
df = df[df['size'] <= 300]
df = df[df['size'] >= 20]
```

```python
df['year'] = pd.to_datetime(df['soldDate']).dt.year
df['month'] = pd.to_datetime(df['soldDate']).dt.month
df['quarter'] = pd.to_datetime(df['soldDate']).dt.quarter
df.head()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Convert lat/lon to radians
coords = np.radians(df[['latitude', 'longitude']].values)

# Earth radius in km
kms_per_radian = 6371.0088

# Function to apply DBSCAN for a given radius (in meters)
def run_dbscan(eps_meters):
    eps = eps_meters / 1000 / kms_per_radian
    db = DBSCAN(eps=eps, min_samples=1, algorithm='ball_tree', metric='haversine')
    return db.fit_predict(coords)

# Run DBSCAN for both radii
df['cluster_100m'] = run_dbscan(100)
df['cluster_1000m'] = run_dbscan(1000)
```
